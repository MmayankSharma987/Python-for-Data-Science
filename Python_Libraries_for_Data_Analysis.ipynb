##1.5) Modifying entries in Numpy array- matrix

```python
rand_mat1 = np.random.randn(5,5)
print('The values before modifications:\n',rand_mat1)
rand_mat1[1:4,1:4]=5 #modifying value to 5 from rows 2,3 and 4 (index-1,2 and 3), column 2,3 and 4 (index-1 to 4), first index value is included while last index value is excluded.
print('The values after modifications:\n',rand_mat1)
print('The values of row 1 and 2, column 2 and 3 are:\n',rand_mat1[0:2,1:3]) #accesing multiple entries- row-1(index 0), row-2(index 1), column-2(index 1), column-3(index 2), in [0:2,1:3] 0 and 1 index is included and 2 and 3 index is excluded.
```

```python
# creating sub-matrix from main matrix, making sub-matrix of first 3 rows
sub_matrix1 = rand_mat1[0:3,0:5]
print('The unmodified sub-matrix1 created from main matrix:\n',sub_matrix1)
# Modify all the values to 3 in sub-matrix1
sub_matrix1[:] = 3
print('The modified sub-matrix1 created from main matrix:\n',sub_matrix1)
```

```python
rand_mat1 # main matrix also got changed once we modify the sub-matrix!
```

```python
# to avoid such changes we have to apply .copy() method when we assign sub-matrix
random_matrix2= np.random.randn(6,6)
print('The raw data of random matrix2:\n',random_matrix2)
sub_matrix1= random_matrix2[0:3,0:7].copy() # making random_matrix2 (main matrix) unmodified
print('The raw data of sub-matrix1:\n',sub_matrix1)
sub_matrix1[:]=5
print('The modified sub-matrix1:\n',sub_matrix1)
print('The main matrix:\n',random_matrix2)
```

##2.1) Pandas - Series and DataFrames
Panda Series is a one dimensional labeled array or list capable of holding any type of data (integer,float,string,boolean,python objects etc). The labels are collectively called as index. Panda series can be thought as a single column of excel spreadsheet and each entry in a series corresponds to an individual row in the spreadsheet. Pandas is used to create and manipulate data sets. It create a data sets or series like NumPy series(array and matrix). Its like an excel sheet in python.

```python
#Panda Series
#creating list of prices of different medicines
med_price_list = [100,200,250,600,800]
print('Med_price_list: ', med_price_list)

# converting med_price_list to array
med_price_arr = np.array(med_price_list)
print('Med_price_array: ', med_price_arr)

# converting list and array into Pandas Series objects
pd_series_list = pd.Series(med_price_list)
print('pd_series_list:\n', pd_series_list)
pd_series_arr = pd.Series(med_price_arr)
print('pd_series_arr:\n', pd_series_arr)
type(pd_series_list)
```

```python
# genertaing strings data type in pd series
med_price_list_labeled = pd.Series(med_price_list, index = ['Dolo650','Cetrizine','Augmentin','Ibuprofin','Pantop'])
print('med_price_list_labeled:\n', med_price_list_labeled)
```

```python
# adding 2.5 to existing price list
updated_med_price_list_labeled = med_price_list_labeled + 2.5
print(updated_med_price_list_labeled)
```

```python
# calculate the difference between two price list
new_price_list = [120,250,300,680,900]
new_med_price_list_labeled = pd.Series(new_price_list,index = ['Dolo650','Cetrizine','Augmentin','Ibuprofin','Pantop'])
print('new_med_price_list_labeled:\n', new_med_price_list_labeled)
print('Difference between the prices of medicine-\n',new_med_price_list_labeled - updated_med_price_list_labeled)
new_med_price_list_labeled[0] # indexing
```

##2.5) Pandas combining DataFrame by Concat,Join and Merge

```python
# practice
import pandas as pd
data_cust= pd.DataFrame({'CustomerID':[101,102,103,103],
                         'Category':['medium','medium','high','low'],
                         'first visit':['yes','no','yes','yes'],
                         'sales':[123,52,214,663],
                         'index':[0,1,2,3]})
data_cust
```

```python
# to calculate measures of central tendency (mean,mode,median,standard deviation) in sales column
data_cust['sales'].mean()
print('mean:',data_cust['sales'].mean())
data_cust['sales'].median()
print('median:', data_cust['sales'].median())
data_cust['sales'].mode()
print('mode:',data_cust['sales'].mode())
data_cust['sales'].std()
print('standard deviation:', data_cust['sales'].std())
data_cust['sales'].sum()
print('sum:', data_cust['sales'].sum)
```

```python
data_cust['Location'] = ['Chicago','London','Boston','Tokyo']
data_cust
```

```python
data_cust.reset_index(drop=True,inplace=True)
data_cust
```

```python
data_cust.sort_values('sales',ascending= True)
```

```python
data_cust= pd.DataFrame({'CustomerID':['101','102','103','104'],
                         'Category':['medium','medium','high','low'],
                         'first visit':['yes','no','yes','yes'],
                         'sales':[123,52,214,663]},
                         index=[0,1,2,3])
data_cust_new = pd.DataFrame({'CustomerID': ['101','103','104','105'],
                              'distance':[12,9,44,21],
                              'sales':[123,214,663,331]},
                             index=[4,5,6,7])
```

```python
data_cust
```

```python
data_cust_new
```

```python
# 1.concat
pd.concat([data_cust,data_cust_new])
```

```python
pd.concat([data_cust,data_cust_new], axis=0)
```

```python
pd.concat([data_cust,data_cust_new],axis=1)
```

```python
#2. Merge- Merge can be done by union (outer) and intersection (inner), merge is based on column
pd.merge(data_cust,data_cust_new, how = 'outer', on = 'CustomerID') # Outer merge is union of on
```

```python
#2. Merge- Merge can be done by union (outer) and intersection (inner)
pd.merge(data_cust,data_cust_new, how = 'inner', on = 'CustomerID') # Inner merge is intersection of on
```

```python
# merge on the basis of precceding (left) or suceeding (right) data frame
pd.merge(data_cust,data_cust_new,how = 'left',on='CustomerID')
```

```python
# merge on the basis of precceding (left) or suceeding (right) data frame
pd.merge(data_cust,data_cust_new,how = 'right',on='CustomerID')
```

```python
# Join- it will join two datafarme on the basis of index value
df1=pd.DataFrame({'Q1':[100,200,300],'Q2':[200,200,100]},index= [0,1,2])
df2=pd.DataFrame({'Q3':[100,200,300],'Q4':[101,201,301]},index= [0,2,3])
```

```python
df1
```

```python
df2
```

```python
df2.join(df1, how ='right'), #outer,inner,left and right works same as merge
```
